# LOC and ILOC

- **Line of Code**
- **Integer Line of Code**

In pandas loc and iloc are two powerfull methods used for selecting data and dataframe

- loc based indexing . it is  used to access rows and columns by there label names.
- when slicing with loc, both start and end lables inclusive

use loc when want to select data by specific 

```python
brics
```

```python
brics.loc[:,:]
```

```python
brics.loc[:,['country']]
```

```python
brics.loc[:,['capital']]
```

```python
brics.loc["country":"capital"]
```

```python
brics.loc[:,"country":"capital"]
```

```python
brics.loc[4,["country","capital"]]
```

```python
brics.loc[1:4,["country","capital"]]
```

```python
brics.loc[1,:]
```

**ILOC**

- Integer based indexing
- used to access rows and columns by their integer position instead of giving the label.

- end is exclusive here



```python
brics
```

```python
brics.iloc[:,:]
```

```python
brics.iloc[:,0]
```

```python
brics.iloc[:,1]
```

```python
brics.iloc[1:4,0]
```

```python
df.head()
```

```python
df.iloc[:,[0,3]]  # only 0th and 3rd col >> If want from 0 to 3rd col use [0:4]
```

```python
df.loc[:,"PassengerId"]
```

```python
df.iloc[:, [3, 5]]
```

```python
df.iloc[0:3, [3, 5]]
```

```python
df.iloc[1:3, [3, 5]]
```
